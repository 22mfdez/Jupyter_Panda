Temas a realizar.
1. Muestra un gráfico de barras con el total de clientes por categoria
2. Muestra un gráfico circular con el precio medio por categoria
3. Muestra un histograma con los precios de los productos


```python
import pandas as pd
import matplotlib.pyplot as plt

datos=pd.read_csv('products.csv')
#print(datos)
df=datos[['unitsInStock', 'categoryID']]
#print(df)
df.head(25).plot.barh(x='unitsInStock', y='categoryID') #CAMPO NUMÉRICO SIEMPRE EN X
plt.show()
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
datos=pd.read_csv('products.csv')
#print(datos)
df=datos[['unitPrice', 'categoryID']]
df.unitPrice.head(10).value_counts().plot.pie()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
datos=pd.read_csv('products.csv')
#print(datos)
df=datos[['unitPrice', 'productName']]
df.unitPrice.plot.hist() #las X SIEMPRE SON VALORES NUMÉRICOS
plt.show()
```
